# 4.On average which region has the highest

a. GDP per capita (current US$)

```python
highest_gdp=data.groupby('Region')['GDP per capita (current US$)'].mean()
print('The region with highest GDP is: ',highest_gdp.nlargest(1))
fig= plt.subplots(figsize=(10,5))
plt.xticks(rotation=45)
ax=sns.lineplot(data=highest_gdp,markers=True)
ax.set(ylabel='Mean of GDP per capita')
```

b. International trade: Imports (million US$)

```python
highest_trade=data.groupby('Region')['International trade: Imports (million US$)'].mean()
print('The region with highest import trade is: ',highest_trade.nlargest(1))
fig= plt.subplots(figsize=(12,4))
plt.xticks(rotation=90)
ax=sns.lineplot(data=highest_trade,markers=True)
ax.set(ylabel='Mean of International Trade: Imports')
```

# 5. Find out which region has more consistent with respect to GDP per capita (currentUS$)

```python
gdp_variation=data.groupby('Region')['GDP per capita (current US$)'].std().round(3)
gdp_variation.sort_values(ascending=True)
print('The region which is the most consistent with GDP is: ',gdp_variation.nsmallest(1))
fig= plt.subplots(figsize=(12,4))
plt.xticks(rotation=90)
ax=sns.lineplot(data=gdp_variation,markers=True)
ax.set(ylabel='Consistant GDP per Capita')
```
