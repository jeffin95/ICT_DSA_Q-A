```python
no_students = len(sd.index)
no_features = len(sd.columns)-1
no_passed = len(sd[(sd.passed == 'yes')])
no_failed = len(sd[(sd.passed == 'no')])
pass_per = float(no_passed)/float(no_students)*100.0
print ("Total number of students: {}".format(no_students))
print ("Number of features: {}".format(no_features))
print ("Number of students who passed: {}".format(no_passed))
print ("Number of students who failed: {}".format(no_failed))
print ("Graduation rate of the class: {:.2f}%".format(pass_per))
```

```python
for i in sd.columns:
    print(i,sd[i].unique())
```

```python
x = sd.drop(['school','age','address','famsize','Medu','Fedu','reason','guardian','paid','nursery',
            'famrel','passed'],axis=1)
y = sd['passed']
```

```python
x.head()
```

```python
y.head()
```

```python
x=pd.get_dummies(x,columns=['activities','sex','Pstatus','famsup','higher','internet','romantic'])
```

```python
y=pd.get_dummies(y,columns=['passed'])
```

```python
x.head()
```

```python
y.head()
```

```python
x['freetimes']=x['freetime']+x['goout']
```

```python
x.head()
```

```python
x=x.drop(['freetime','goout','Mjob','Fjob','absences'],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import f1_score,confusion_matrix
print('f1_score:',f1_score(y_test,y_pred,average='weighted'))
print('confusion_matrix:',confusion_matrix(y_test.values.argmax(axis=1),y_pred.argmax(axis=1)))
```

```python
pd.Series(rf.feature_importances_,index=x.columns).sort_values(ascending = False)*100
```

```python
rf.fit(x_train,y_train)
rft= RandomForestClassifier(n_estimators=800,max_depth=10,random_state=42,criterion="entropy")
rft.fit(x_train,y_train)
y_pred=rft.predict(x_test)
f1_score(y_test,y_pred,average='weighted')
```
